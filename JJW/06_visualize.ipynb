### seaborn을 이용한 시각화

```python
# seaborn으로 시계열 데이터 분석하기
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../DATA/data_with_btc_scaled.csv')
data
```

```python
# 시계열 데이터를 그래프로 그리기
plt.figure(figsize=(16, 8))
sns.lineplot(data=data, palette='tab10', linewidth=2.5)
plt.show()
```

```python
# 그냥 plt
plt.figure(figsize=(16, 8))
plt.plot(data['btc'], label='BTC', color='orange', linewidth=2, alpha=1)
plt.plot(data['Gold'], label='Gold', color='gold', linewidth=1, alpha=1)
plt.plot(data['Dollar'], label='Dollar(DXY)', color='green', linewidth=1, alpha=0.7)
plt.plot(data['SnP'], label='S&P 500', color='skyblue', linewidth=1, alpha=1)
plt.legend()
plt.xticks(range(0, data.shape[0], 300), data['date'].loc[::300], rotation=0)
plt.title('BTC, Gold, DXY, S&P 500')
plt.show()
```
