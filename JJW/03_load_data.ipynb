### DB 접속해 데이터 불러오기

```python
from sqlalchemy import Column, Integer, String, Date, Float, Table, Select, Double
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy.orm import sessionmaker
from sqlalchemy import create_engine
import csv

# Define the base model class
Base = declarative_base()



# Create the engine and session
engine = create_engine('mysql+pymysql://root:kdt5@1.251.203.204:33065/Team2')
Session = sessionmaker(bind=engine)
session = Session()
```

```python
# 테이블과 연결
table_name = 'BTCUSDT_1d'  # 테이블 이름
table = Table(table_name, Base.metadata, autoload_with=engine)

# 쿼리 생성
query = Select(table.c.timestamp, table.c.close)

# 쿼리 실행
result = session.execute(query)
```

```python
print(result.fetchall())
```

```python
# dataframe으로 변환
import pandas as pd
df = pd.DataFrame(result.fetchall(), columns=result.keys())
print(df)
```

```python
# csv로 저장
df.to_csv('BTCUSDT_1d.csv', index='timestamp')

# Close the session
session.close()
```
